```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import xarray as xr
import h5py
from tqdm import tqdm
import time

from APLB_acc_NoSTC_3D import APLB_acc_NoSTC_3D
from APLB_acc_NoSTCApril_3D import APLB_acc_NoSTCApril_3D
from APLB_acc_anySTC_3D import APLB_acc_anySTC_3D
from APLB_acc_LCApril_3D import APLB_acc_LCApril_3D
```

```python
c = 2.9979e8
lambda_0 = 1800e-9  # central wavelength
omega_0 = 2*np.pi*c/lambda_0

s = 57.5
tau_FWHM = np.sqrt(2)*s*np.sqrt(2**(2/(s+1))-1)/omega_0
tau_0 = s*np.sqrt(np.exp(2/(s+1))-1)/omega_0
# print(tau_0)

NA = 0.1
half_angle = np.arcsin(NA)
a = 2/(np.sin(half_angle)*np.tan(half_angle)*2*np.pi/lambda_0)  # set confocal parameter
w_0 = 2/(np.tan(half_angle)*2*np.pi/lambda_0)  # set beam waist
# print('{0:.3g}'.format(1e6*w_0))

P = 100e12  # FL power
t_0 = -10*tau_0

x_0 = 0.2*w_0
y_0 = 0.2*w_0
```

```python
t, z, x, y, v_z, v_x, v_y, KE = APLB_acc_NoSTC_3D(lambda_0, tau_0, w_0, P, 0.0, 0.0, t_0, 0.0, x_0, y_0, 0.0)
t_April, z, x, y, v_z, v_x, v_y, KE_April = APLB_acc_NoSTCApril_3D(lambda_0, s, a, P, 0.0, t_0, 0.0, x_0, y_0, 0.0)
t_anySTC, z, x, y, v_z, v_x, v_y, KE_anySTC = APLB_acc_anySTC_3D(lambda_0, tau_0, w_0, P, 0.0, np.array([0.0]), np.array([0.0]), np.array([0.0]), 0.0, 0.0, x_0, y_0, 0.0)
t_LCApril, z, x, y, v_z, v_x, v_y, KE_LCApril = APLB_acc_LCApril_3D(lambda_0, s, a, P, 0.0, 0.0, 0.0, t_0, 0.0, x_0, y_0, 0.0, 0.0)
```

```python
plt.plot(t, KE)
plt.plot(t_April, KE_April)
plt.plot(t_anySTC, KE_anySTC)
plt.plot(t_LCApril, KE_LCApril)
```

```python
t_April, z, x, y, v_z, v_x, v_y, KE_April = APLB_acc_NoSTCApril_3D(lambda_0, s, a, P, 0.0, t_0, 0.0, x_0, y_0, 0.0)
t_LCApril, z, x, y, v_z, v_x, v_y, KE_LCApril = APLB_acc_LCApril_3D(lambda_0, s, a, P, 0.0, 0.0, 0.0, t_0, 0.0, x_0, y_0, 0.0, tau_0/50)
```

```python
plt.plot(t_April, KE_April)
plt.plot(t_LCApril, KE_LCApril)
plt.xlim(-1e-13, 5e-13)
```
